```python
# import das bibliotecas:
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Criacao do dataframe1 aqui:
pd1 = pd.read_csv('Mod0',delim_whitespace=True,skiprows=2)
```

```python
# visualização do dataframe para validação
pd1
```

```python
# Plot aqui: 
plt.figure(figsize=(18,8),facecolor='w')

p1 = plt.scatter(pd1['Long'],pd1['Lat'],c = pd1['F'],cmap='jet')
cbar = plt.colorbar(p1, aspect=30, fraction = 0.03, orientation="horizontal")
cbar.set_label('F $ (nT) $',fontsize=15,rotation = 0)
plt.xlabel('Long', fontsize=16)
plt.ylabel('Lat ', fontsize=16)
plt.title('nT',fontsize=20)
plt.grid()

plt.show()
```

### Etapa 1: Definicão da amostragem de pontos (UTMX, UTMY) a serem criados:


```python
# Codigo aqui:
xmin = pd1.min()['Long']
xmax = pd1.max()['Long']
x = np.linspace(xmin, xmax, 50)

ymin = pd1.min()['Lat']
ymax = pd1.max()['Lat']
y = np.linspace(ymin, ymax, 50)

```

```python
# comandos aqui:
X,Y = np.meshgrid(x,y)
```

```python
# plot dos pontos interpolados aqui:
plt.figure(figsize=(12,7), facecolor='w')
p1 = plt.scatter(pd1['Long'], pd1['Lat'], c=pd1['F'])
cbar = plt.colorbar(p1, aspect=30, fraction = 0.03, orientation="horizontal")
cbar.set_label('F (nT)',fontsize=15,rotation = 0)
plt.xlabel('Long', fontsize=16)
plt.ylabel('Lat', fontsize=16)
plt.grid()

p2 = plt.scatter(X, Y, s=10.0, c='k', marker='*') 
plt.xlabel('Long ', fontsize=16)
plt.ylabel('Lat ', fontsize=16)
plt.grid()
plt.show()
```

```python
# importar biblioteca aqui:
from scipy.interpolate import griddata 
```

```python
# criacao de arrays para facilitar a funcao de interpolacao (SENAO DÁ CHABUUUU!!!! - limitação do pandas!):
long = np.array(pd1['Long'])
lat = np.array(pd1['Lat'])
F = np.array(pd1['F'])
#std  = np.array(pd1['STD'])
```

```python
# interpolacao aqui:
Fi_linear  = griddata( (long, lat), F, (X,Y), method='linear')
Fi_cubic   = griddata( (long, lat), F, (X,Y), method='cubic')
```

```python
#Plot para verificação (através da funcao plt.scatter)
#Gráficos aqui:
plt.figure(figsize=(18,8),facecolor='w')

plt.subplot(121)
plt.title('Linear', fontsize=16)
p1 = plt.scatter(pd1['Long'],pd1['Lat'], c=pd1['F'],cmap='jet')
p2 = plt.scatter(X, Y, s=10.0, c=Fi_linear,cmap='hsv') 
cbar = plt.colorbar(p2, aspect=30, fraction = 0.03, orientation="horizontal")
cbar.set_label('F $(nT)$',fontsize=15,rotation = 0)
plt.xlabel('Long', fontsize=16)
plt.ylabel('Lat', fontsize=16)

plt.subplot(122)
plt.title('Cubic', fontsize=16)
p1 = plt.scatter(pd1['Long'],pd1['Lat'], c=pd1['F'],cmap='jet')
p2 = plt.scatter(X, Y, s=10.0, c=Fi_cubic,cmap='hsv') 
cbar = plt.colorbar(p2, aspect=30, fraction = 0.03, orientation="horizontal")
cbar.set_label('F $(nT)$',fontsize=15,rotation = 0)
plt.xlabel('Long', fontsize=16)
plt.ylabel('Lat' , fontsize=16)

plt.show()
```

```python
#Usando as funções de mapa - plt.contourf() para criação dos mapas 
#Gráficos aqui:

plt.figure(figsize=(17,7), facecolor='w')

plt.subplot(121)
p2 = plt.contourf(X, Y, Fi_linear, 5,cmap = 'Accent') 
cbar = plt.colorbar(p2, aspect=10, fraction = 0.03, orientation="horizontal")
cbar.set_label('F $(nT)$',fontsize=15,rotation = 0)
plt.xlabel('Long', fontsize=16)
plt.ylabel('UTM-Y($m$)', fontsize=16)
plt.grid()

plt.subplot(122)
p2 = plt.contourf(X, Y, Fi_cubic, 300,cmap='jet') 
cbar = plt.colorbar(p2, aspect=30, fraction = 0.03, orientation="horizontal")
cbar.set_label('F $(nT)$',fontsize=15,rotation = 0)
plt.xlabel('Long', fontsize=16)
plt.ylabel('Lat', fontsize=16)
plt.grid()

plt.show()
```

```python
# calculo do mapa residual aqui:
res = Fi_cubic - Fi_linear
```

```python
plt.figure(figsize=(17,7), facecolor='w')
plt.title('Residuals', fontsize=16)
p2 = plt.contourf(X, Y, res, 30, cmap='jet') 
cbar = plt.colorbar(p2, aspect=30, fraction = 0.03, orientation="horizontal")
cbar.set_label('F residual $(nT)$',fontsize=15,rotation = 0)
plt.xlabel('Long', fontsize=16)
plt.ylabel('Lat', fontsize=16)
plt.grid()
```
